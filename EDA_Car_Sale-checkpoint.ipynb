The source of data is from github website.This dataset contains data for more than 9.5K cars sales in Ukraine. https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/car_sales.csv

Dataset contains 9576 rows and 10 variables with essential meanings:

-	car: manufacturer brand
-	price: seller’s price in advertisement (in USD)
-	body: car body type
-	mileage: of car( km) as it is used ones
-	engV: rounded engine volume (‘000 cubic cm)
-	engType: type of fuel (“Other” in this case should be treated as NA)
-	registration: whether car registered in Ukraine or not

-	model: specific model name

# 4.1. What are the average prices of cars with type,year of make,engine power and engine type?
<a id=section401></a>

```python
df.groupby(['car','body','year','engV','engType'])['price'].mean()
```

```python
trial = pd.DataFrame()
for b in list(df["car"].unique()):
    for v in list(df["body"].unique()):
        z = df[(df["car"] == b) & (df["body"] == v)]["price"].mean()
        trial = trial.append(pd.DataFrame({'car':b , 'body':v , 'avgPrice':z}, index=[0]))
trial = trial.reset_index()
del trial["index"]
trial["avgPrice"].fillna(0,inplace=True)
trial["avgPrice"].isnull().value_counts()
trial["avgPrice"] = trial["avgPrice"].astype(int)
```

```python
# HeatMap tp show average prices of vehicles by brand and type together
tri = trial.pivot("car","body", "avgPrice")
fig, ax = plt.subplots(figsize=(15,20))
sns.heatmap(tri,linewidths=1,cmap="YlGnBu",annot=True, ax=ax, fmt="d")
ax.set_title("Average price of vehicles by vehicle type and brand",fontdict={'size':20})
ax.xaxis.set_label_text("Type Of Vehicle",fontdict= {'size':20})
ax.yaxis.set_label_text("Brand",fontdict= {'size':20})
plt.show()
```

```python
#function to plot heatmap
def heat_map(trial):
    hmap= trial.pivot("car","body", "avgPrice")
    fig, ax = plt.subplots(figsize=(15,20))
    sns.heatmap(hmap,linewidths=1,cmap="YlGnBu",annot=True, ax=ax, fmt="d")
    ax.set_title("Average prices of top car brands with vehicle type",fontdict={'size':20})
    ax.xaxis.set_label_text("Type Of Vehicle",fontdict= {'size':20})
    ax.yaxis.set_label_text("car",fontdict= {'size':20})
    plt.show()
    return None  
```

```python
def top_brand_affordable(most_popular_brands):
    #displaying top  brands with their mean mileage and mean price 
    brands_mean_price = {}
    global mean_mileage
    global mean_price
   
    for brands in most_popular_brands:
      
        brand_only = affordable_cars[affordable_cars['car'] == brands]
        mean_price = brand_only['price'].mean()
        brands_mean_price[brands] = int(mean_price)
        
    brand_mean_mileage = {}
    
    for brands in most_popular_brands:
        brand_only = affordable_cars[affordable_cars['car'] == brands]
        mean_mileage = brand_only['mileage'].mean()
        brand_mean_mileage[brands] = int(mean_mileage)
        
    mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending = False)
    mean_price = pd.Series(brands_mean_price).sort_values(ascending = False)
    
    top_brand_info = pd.DataFrame(mean_mileage, columns = ['mean_mileage'])
    top_brand_info['mean_price'] = mean_price
    
    
    return top_brand_info
```

```python
def top_brand_expensive(most_popular_brands):
    #displaying top  brands with their mean mileage and mean price 
    brands_mean_price = {}
    global mean_mileage
    global mean_price
  
   
    for brands in most_popular_brands:
      
        brand_only = expensive_cars[expensive_cars['car'] == brands]
        mean_price = brand_only['price'].mean()
        brands_mean_price[brands] = int(mean_price)
        
    brand_mean_mileage = {}
    
    for brands in most_popular_brands:
        brand_only = expensive_cars[expensive_cars['car'] == brands]
        mean_mileage = brand_only['mileage'].mean()
        brand_mean_mileage[brands] = int(mean_mileage)
        
    mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending = False)
    mean_price = pd.Series(brands_mean_price).sort_values(ascending = False)
    
    top_brand_info = pd.DataFrame(mean_mileage, columns = ['mean_mileage'])
    top_brand_info['mean_price'] = mean_price
    
    
    return top_brand_info
```

```python
def top_brand_luxury(most_popular_brands):
    #displaying top  brands with their mean mileage and mean price 
    brands_mean_price = {}
    global mean_mileage
    global mean_price
   
    for brands in most_popular_brands:
      
        brand_only = luxury_cars[luxury_cars['car'] == brands]
        mean_price = brand_only['price'].mean()
        brands_mean_price[brands] = int(mean_price)
        
    brand_mean_mileage = {}
    
    for brands in most_popular_brands:
        brand_only = luxury_cars[luxury_cars['car'] == brands]
        mean_mileage = brand_only['mileage'].mean()
        brand_mean_mileage[brands] = int(mean_mileage)
        
    mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending = False)
    mean_price = pd.Series(brands_mean_price).sort_values(ascending = False)
    
    top_brand_info = pd.DataFrame(mean_mileage, columns = ['mean_mileage'])
    top_brand_info['mean_price'] = mean_price
    
    
    return top_brand_info
```

```python
def draw_donut(most_popular_brands):
    #Function to create donut chart
    df1 = pd.DataFrame({'mean_mileage':mean_mileage,'mean_price':mean_price})
    df1.index = most_popular_brands
    
    names=most_popular_brands
    mile=df1.mean_mileage
    price1=df1.mean_price
    
    def absolute_value(val):
        a  = np.round(val/100.*mile.sum(), 0)
       
        return a
     
    def absolute_value1(val):
        a  = np.round(val/100.*price1.sum(), 0)
        
        return a
    
    # Create colors
    a, b, c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]
    my_circle=plt.Circle( (0,0), 0.9, color='white')
 
    # First Ring (outside)
    fig, ax = plt.subplots()
    ax.axis('equal')

    # Top cars
    plt.pie(mile, labels=names,autopct =  absolute_value,radius=1.5, colors = ['yellowgreen', 'gold','green','blue'])
    #mean mileage
    plt.pie(mile,labels=names,radius=1.5, colors = ['yellowgreen', 'gold','green','blue'])
    plt.pie(mile,radius=1.5-0.3, colors=[a(0.5)])
   
    #mean price

    plt.pie(price1,autopct = absolute_value1,radius=1.2-0.5,colors = [b(0.4)])
    plt.pie(price1,radius=0.7-0.4,colors = ['white', 'white'])

    
    plt.margins(0,0)
    #plt.title('Donut plot')
    plt.show()
    
    return None

    
    
```

```python
trial = pd.DataFrame()
for b in list(most_popular_brands):
    for v in list(most_popular_cartypes):
        z = affordable_cars[(affordable_cars["car"] == b) & (affordable_cars["body"] == v)]["price"].mean()
        trial = trial.append(pd.DataFrame({'car':b , 'body':v , 'avgPrice':z}, index=[0]))
trial = trial.reset_index()
del trial["index"]
trial["avgPrice"].fillna(0,inplace=True)
trial["avgPrice"].isnull().value_counts()
trial["avgPrice"] = trial["avgPrice"].astype(int)
heat_map(trial)
```
